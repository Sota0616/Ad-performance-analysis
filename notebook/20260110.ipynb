```python
import pandas as pd

df = pd.read_csv("final_shop_6modata.csv")
df.head()
```

```python
df.info()
```

```python
df["CTR_calc"] = df["Clicks"] / df["Impressions"]
df["CPC_calc"] = df["Cost"] / df["Clicks"]
df["CVR_calc"] = df["Conversions"] / df["Clicks"]
```

```python
ad_summary = (
    df.groupby("Ad Group")
      .agg(
          impressions=("Impressions","sum"),
          clicks=("Clicks","sum"),
          conversions=("Conversions","sum"),
          cost=("Cost","sum"),
          revenue=("Revenue","sum"),
          pnl=("P&L","sum")
      )
)

ad_summary["CTR"] = ad_summary["clicks"] / ad_summary["impressions"]
ad_summary["CVR"] = ad_summary["conversions"] / ad_summary["clicks"]
ad_summary["CPA"] = ad_summary["cost"] / ad_summary["conversions"]
```

```python
#平均購入単価が出る。
#CPA < 平均購入単価 → 生存
#CPA > 平均購入単価 → 赤字構造

ad_summary["revenue"].sum() / ad_summary["conversions"].sum()
```

```python
#平均購入単価である5.85を超えるデータだけを算出
bad_cpa = ad_summary[ad_summary["CPA"] > 5.85]
bad_cpa[["CPA", "pnl"]].sort_values("CPA", ascending=False)
```

```python
#それぞれのデータ要素を列挙
bad_cpa[["CPA", "CTR", "CVR", "pnl"]]
```

```python
#データ解釈のために全体の平均データを算出
avg = ad_summary[["CTR", "CVR", "CPA"]].mean()
avg
```

```python
bad_ctr = bad_cpa[bad_cpa["CTR"] < 0.25]
bad_cvr = bad_cpa[bad_cpa["CVR"] <  0.11]
```

```python
bad_ctr_only = bad_ctr[~bad_ctr.index.isin(bad_cvr.index)]
bad_cvr_only = bad_cvr[~bad_cvr.index.isin(bad_ctr.index)]
bad_both = bad_ctr.index.intersection(bad_cvr.index)
```
